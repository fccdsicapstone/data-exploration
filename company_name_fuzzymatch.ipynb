## Company Name Match

### Used the fuzzywuzzy pacakge 
### Token_set_ratio() function with 95 score to make sure the matches are correct 
### Count the number of unmatched items in each of the name column in fcc
### manually compared and replaced the unmatched items 

```python
from fuzzywuzzy import fuzz
import pandas as pd
```

provider = pd.read_csv('provider_name.csv')
holding_company = pd.read_csv('holding_company_name.csv')
holding_final = pd.read_csv('holding_final.csv')
fund_company_name = pd.read_csv('fund_company_cleaned.csv')

```python
provider_match = []
holding_company_match = []
holding_final_match = []
provider_none = 0
holding_company_none = 0
holding_final_none = 0
for i in fund_company_name['company_name']:
    provider_ratio=[]
    holding_company_ratio=[]
    holding_final_ratio = []
    provider_sort_ratio=[]
    holding_company_sort_ratio=[]
    holding_final_sort_ratio = []
    for j in provider['provider_name']:
        provider_ratio.append(fuzz.token_set_ratio(i, j))
        provider_sort_ratio.append(fuzz.token_sort_ratio(i, j))
    if (max(provider_sort_ratio) > 95):
        relevant_ind = provider_sort_ratio.index(max(provider_sort_ratio))
        provider_match.append(provider['provider_name'][relevant_ind])
    else:
        provider_match.append('none')
        provider_none +=1
        
    for k in holding_company['holding_company_name']:
        holding_company_ratio.append(fuzz.token_set_ratio(i, k))
        holding_company_sort_ratio.append(fuzz.token_sort_ratio(i, k))
    if (max(holding_company_sort_ratio)>95):
        relevant_ind = holding_company_sort_ratio.index(max(holding_company_sort_ratio))
        holding_company_match.append(holding_company['holding_company_name'][relevant_ind])
    else:
        holding_company_match.append('none')
        holding_company_none +=1
        
    for m in holding_final['hoco_final']:
        holding_final_ratio.append(fuzz.token_set_ratio(i, m))
        holding_final_sort_ratio.append(fuzz.token_sort_ratio(i, m))
    if (max(holding_final_sort_ratio) > 95):
        relevant_ind = holding_final_sort_ratio.index(max(holding_final_sort_ratio))
        holding_final_match.append(holding_final['hoco_final'][relevant_ind])
    else:
        holding_final_match.append('none')
        holding_final_none += 1
```

```python
provider_match_df = pd.DataFrame(provider_match, columns = ['provider_matched'])
holding_company_match_df = pd.DataFrame(holding_company_match, columns = ['holding_company_matched'])
holding_final_df = pd.DataFrame(holding_final_match, columns = ['holding_final_matched'])
```

```python
temp = pd.concat([fund_company_name, provider_match_df, holding_company_match_df, holding_final_df], axis=1)
temp
```
